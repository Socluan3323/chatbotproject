```python
class Boiler(Node):
    def prep(self,shared):
        return shared.get("water_temperature")
    def exec(self,prep_res):
        C_degree_change = 100 - prep_res
        temp_after_boiled = 100
        return temp_after_boiled, C_degree_change
    def post(self,shared,prep_res,exec_res):
        water_temperature, C_degree_change = exec_res
        shared["water_temperature"] = water_temperature
        shared["degree_increase"] = C_degree_change
        if shared.get("history",None):
            shared["history"].append(("increase",C_degree_change))
        else:
            shared["history"] = []
            shared["history"].append(("increase",C_degree_change))
        return None



class Freeze(Node):
    def prep(self,shared):
        return shared.get("water_temperature")
    def exec(self,prep_res):
        if prep_res >= 50:
            temp_after_freeze = prep_res //2
            degree_drop = prep_res - temp_after_freeze
        else:
            temp_after_freeze = prep_res - 10
            degree_drop = prep_res - temp_after_freeze
        return temp_after_freeze, degree_drop
    def post(self,shared,prep_res,exec_res):
        temp_after_freeze, degree_drop = exec_res
        shared["water_temperature"] = temp_after_freeze
        shared["degree_drop"] = degree_drop
        if shared.get("history",None):
            shared["history"].append(("drop",degree_drop))
        else:
            shared["history"] = []
            shared["history"].append(("drop",degree_drop))
        return None
    
```

```python
p = {
    'water_temperature' : 80


}
f = Freeze()
b = Boiler()



```

```python
b.run(p)
f.run(p)
print(p)
```

```python
shared = {"user_query": "xin chao",
          "bot_response": "",
          }
```

```python
class bot(Node):
    def prep(self,shared):
        return shared["user_query"]
    def exec(self,prep_res):
        user_query = prep_res
        return call_llm(user_query)
    def post(self, shared, prep_res, exec_res):
        shared["bot_response"] = exec_res

```

```python
p = bot()
p.run(shared)
```
